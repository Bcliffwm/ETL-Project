```python
population_cols = ["Geography", "Population Estimate (as of July 1) - 2014", "Population Estimate (as of July 1) - 2015",
                  "Population Estimate (as of July 1) - 2016"]
population_transformed = population_df[population_cols].copy()

#Rename Column Headers
population_transformed = population_transformed.rename(columns={"Geography":"State",
                                                               "Population Estimate (as of July 1) - 2014": "2014_pop",
                                                               "Population Estimate (as of July 1) - 2015": "2015_pop",
                                                               "Population Estimate (as of July 1) - 2016": "2016_pop"})
population_transformed.head()
```

```python
homelessness_cols = ["State", "CoC Number", "Measures", "Count"]
homelessness_transformed = homelessness_df[homelessness_cols].copy()

homelessness_transformed = homelessness_transformed.rename(columns={"State":"State_Name"})
homelessness_transformed.head()
```

```python
#Creating database connection
```

```python
connection_string = "postgres:postgres@localhost:5432/ETL_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Confirming Tables
engine.table_names()
```

```python
#Using Pandas to load converted DataFrame into database
```

```python
population_transformed.to_sql(name='population', con=engine, if_exists='append', index=False)
```

```python
#Confirming data has been added by querying the population table
pd.read_sql_query('select * from population', con=engine).head()
```
